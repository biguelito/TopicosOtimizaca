```python
import numpy as np
import plotly.graph_objs as go
```

```python
# Funcao que determina o lucro de um determinado produto y, sendo y
# Ou acucar
# Ou etanol

# s = Tamanho do terreno gasto para produzir cana 
# k = Conversao do terreno em cana de acucar 
# a1y = Conversao da cana de acucar em uma comodite
# cy = Lucro por unidade vendida

def calculate_profit(
    sy: int,
    k: float,
    a1y: float,
    cy: float
):
    return round((cy * ((sy * k) / a1y)), 2) 
```

```python
def sugar_profit(s, k, a1y, cy, qe):
    s2 = s * qe
    return calculate_profit(s2, k, a1y, cy)

def etanol_profit(s, k, a1y, cy, qe):
    s1 = s * (1 - qe)
    return calculate_profit(s1, k, a1y, cy)
```

```python
# Variaveis
# Comodites
y = {
    0: "Cana de acucar",  # Nao sera vendido
    1: "Acucar",
    2: "Etanol"
}
# Tamanho do terreno gasto para produzir cana
s = 500
# Conversao do terreno em cana de acucar
k = 1.2
# Conversao da cana de acucar em uma comodite
a11 =  0.8
a12 =  0.9
# Lucro por unidade vendida
c1 = 1.3
c2 = 1.5

# Numero de casos a testar com terreno parcialmente dedicado para um comodite 
number_cases = 15
# Percentual do terreno dedicado para uma comodite especifica
# Para este, foi escolhido o percentual dedicado para etanol
qe_cases = np.linspace(0, 1, number_cases)
# Tamanho do terreno gasto para produzir cana para um comodite especifico
s1 = s * (1 - qe_cases) 
s2 = s * qe_cases

profits = {
    "proportions": qe_cases,
    "sugar": [sugar_profit(s, k, a11, c1, qe) for qe in qe_cases],
    "etanol": [etanol_profit(s, k, a12, c2, qe) for qe in qe_cases]
}

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=profits["proportions"],
    y=profits["sugar"],
    # mode="lines",
    name="sugar"
)
)
fig.add_trace(go.Scatter(
    x=profits["proportions"],
    y=profits["etanol"],
    # mode="lines",
    name="etanol"
))

fig.update_layout(
    title=f"Lucro de venda",
    xaxis_title="Proporcao da cana em etanol",
    yaxis_title="Lucro",
    legend_title="Comodities",
)

fig
```
